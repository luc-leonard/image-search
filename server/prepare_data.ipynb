```python
from pathlib import Path

import pandas as pd
from fastcore.parallel import parallel
import requests
import urllib.request
from functools import partial
```

```python
files = pd.read_csv('./data/photos.tsv000', delimiter='\t')
```

```python
files
```

```python
def download(row, folder_dest):
    try:
        row = row[1]
        url = row['photo_image_url']
        id = row['photo_id']
        urllib.request.urlretrieve(url, Path(folder_dest) / (id + '.jpg'))
    except Exception as ex:
        ...

download_partial = partial(download, folder_dest='./data/content')
download_partial.__module__ = download.__module__

# download_partial(url='foo.bar', id='42')
parallel(download_partial, list(files.iterrows()), progress=True)
```
